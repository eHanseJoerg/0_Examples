```python
file = '../1_datAcron/data/allflights.csv'
flight_arrivals = []
chunksize = 500

i=0

# we want the extract all flights and their arrival times form the CSV file, then save it into a new, smaller csv.
for dataFrame in pd.read_csv(file, chunksize=chunksize):
    #extract departure, destination, arrival time:
    dataFrame2 = dataFrame['3'].str.split(';', expand=True)
    dataFrame3 = dataFrame2[[0, 1, 2, 3, 4, 85]].copy()
    dataFrame3['last'] = dataFrame3[85].str.split().str[-1].str.slice(0,12)
    dataFrame3 = dataFrame3.drop(85, axis=1)

    #Filter out erroneous data rows:
    mask = (dataFrame3['last'].str.slice(0,4) == '2016')
    dataFrame4 = dataFrame3[mask].copy()
    flight_arrivals.append(dataFrame4)
     
    clear_output(wait=True)
    i= i+1
    print('Iteration ' + str(i) + ', valid rows: ' + str(len(dataFrame4)))


dfflights = pd.concat(flight_arrivals)
dfflights['last'] = pd.to_datetime(dfflights['last'], format='%Y%m%d%H%M')

dfflights.describe()
```
