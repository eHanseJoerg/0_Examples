## Vorsicht! 

Wenn das np.ndarray als ```int64``` instantiiert ist, dann kann es zu Fehlern kommen, wenn man da ein Float - Wert einfügen will! Dann werden die Nachkommastellen ommitted! Um das zu verhindern, sollte man das numpy-Arrray explizit mit dem korrekten Datentyp instantiieren:

````python
a = array([0,1,0])
print (str(type(a)) + str(a.dtype))
````

````python
#Hier passiert der Fehler: statt 10,6 wird 10 im numpy-Array gespeichert.
a[0] = 10.6
a
````

````python
#Bessere Alternative: das Array als float instantiieren:
b = np.array([0, 1, 0], dtype=float)
b
````

````python
b[0] = 10.6
b
````

## Slicing Basics

```
var[lower:upper:step]
```

Eine gute Idee, um über das Slicing nachzudenken ist sich zu überlegen, dass die Indizes zwischen und vor den eigentlichen Array-Elementen sind:

```
# indices:     0   1   2   3   4   5
#              |   |   |   |   |   |
>>> a = arrray([10, 11, 12, 13, 14, 15])
```

Wenn wir dann slicen, z.B. ```a[1:3]```, dann gibt uns das alles, was zwischen den Indizes 1 und 3 ist!

````python
a = array((10, 11, 12, 13, 14, 15))
a[1:3]
````

````python
#omitted boundaries are assumed to be beginning, end of list:
a[:3]
````

````python
#grab the last two elements:
a[-2:]
````

````python
#jedes zweite Element rausziehen:
a[::2]
````

````python
# Kleine Spielerei: Ableitung des Sinus bestimmen "von Hand". 
# Dies ist eine Demo, wie man in numpy vektorwertige Berechnungen durchführt.

x = linspace(0, 2*pi, 101)
y = sin(x)

y_low = y[:-1]
y_upp = y[1:]

x_low = x[:-1]
x_upp = x[1:]

# hier werden die vektorwertigen Berechnungen durchgeführt: statt eines Skalars subtrahieren wir Vektoren gegeneinander:
dy = y_upp - y_low
dx = x_upp - x_low

derivative = dy / dx
derivative
````

````python
plt.plot(y)
plt.plot(cos(x))
plt.plot(derivative)
plt.legend(['x','cos', 'der']);
````

````python
plt.plot(cos(x)[:-1] - derivative);
````
